```python
import cv2 
from PIL import Image, ImageFilter 
import numpy as np 
import cv2
import numpy as np

input_txt = "Ronaldo_Output_1280x720_bin_th=90.txt"
output_image = "Result_Semulation.jpg"

width = 1280
height = 720
expected_size = width * height

# Read pixels
with open(input_txt, "r") as f:
    pixels = [int(line.strip()) for line in f if line.strip()]

print("Received pixels:", len(pixels))
print("Expected pixels:", expected_size)

# Pad missing pixels with 0 (black)
if len(pixels) < expected_size:
    missing = expected_size - len(pixels)
    print(f"Padding {missing} missing pixels with 0")
    pixels.extend([0] * missing)

# If too many pixels (safety)
pixels = pixels[:expected_size]

# Convert to image
img = np.array(pixels, dtype=np.uint8).reshape((height, width))

# Optional: normalize (recommended for Sobel visibility)
img = cv2.normalize(img, None, 0, 255, cv2.NORM_MINMAX)

# Save image
cv2.imwrite(output_image, img)

print("Image saved:", output_image)


```
